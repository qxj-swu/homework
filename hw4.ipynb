## 问题 1 
**最小生成树（Minimum Spanning Tree）**

设  **G**  为一个带权重的连通无向图，且所有边的权重均不相等。令$e_i$ 为权重第 $i$ 小的边。最小生成树（MST）是否必须包含 $e_1$ ? 同理，是否必须包含 $e_2$ 和 $e_3$ ? 若必须包含，请给出证明；否则，请构造反例。需从基本原理论证，不能依赖割引理(cut lemma) 或 Prim/Kruskal算法的正确性。


answer:  
1.任何MST必须包含权重最小的边e₁ 
证明：  
假设存在一个MST T不包含e₁=(u,v)。  
由于T是生成树，u和v在T中必须连通，设P是T中连接u和v的路径。  
因为e₁∉T，所以P中所有边的权重都>w(e₁)（因为e₁是全局最小边）。  
将P中权重最大的边e'替换为e₁，得到新生成树T'=T∪{e₁}\{e'}。  
此时w(T') = w(T) - w(e') + w(e₁) < w(T)，因为w(e₁)<w(e')。  
这与T是MST矛盾。  
因此，任何MST必须包含e₁。  

2.任何MST必须包含权重第二小的边e₂  
证明：    
假设存在一个MST T不包含e₂=(p,q)。  
由于T是生成树，p和q在T中必须连通，设P'是T中连接p和q的路径。  
因为e₂∉T，所以P'中所有边的权重都>w(e₂)（否则会包含更小的边）。  
将P'中权重最大的边e''替换为e₂，得到新生成树T''=T∪{e₂}\{e''}。  
此时w(T'') = w(T) - w(e'') + w(e₂) < w(T)，因为w(e₂)<w(e'')。  
这与T是MST矛盾。  
因此，任何MST必须包含e₂。  

3.存在MST不包含权重第三小的边e₃  
证明：  
考虑图G=(V,E)，其中：  
V = {A,B,C}，E = {e₁=AB(w=1), e₂=BC(w=2), e₃=AC(w=3)}  
可能的生成树：  
T₁ = {AB, BC} (w=1+2=3)，T₂ = {AB, AC} (w=1+3=4)，T₂ = {BC, AC} (w=2+3=5)  
显然T₁是唯一的MST，且不包含e₃。  
因此，MST不一定包含e₃。  

## 问题 2 
**瓶颈生成树（Bottleneck Spanning Tree）**

带有权重的无向图 $G(V,E,w)$ 的瓶颈生成树，表现为：在所有生成树中，最大权重边的权重值最小。即，BST $T$ 最小化瓶颈损失 $c(T)=max_{e \in T}{w(e)}$。

1. 证明 $G$ 的每一个最小生成树（MST）都是瓶颈生成树（BST）
2. 设计一个线性时间复杂度的算法：， 对于一个图 $G(V,E,w)$ 和一个整数 $b$，判断图 $ G$ 是否存在一个瓶颈生成树，其最大权重边的权重不超过 $b$，分析算法设计思路，并基于python编程实现。
3. 设计一个线性时间复杂度的算法：对于给定的图 $G(V,E,w)$，找到其瓶颈生成树，分析算法设计思路，并基于python编程实现。

idea：  
1.证明：
假设存在一个生成树T，其瓶颈值c(T′)<c(T)，其中T是MST，c(T)表示生成树的最大边权。由于T′的最大边权更小，所有T′的边权均 ≤c(T′)。因此T′的边权和sum(T′)≤(n−1)⋅c(T′)。而T的边权和sum(T)≥(n−1)⋅c(T)，因为至少有一条边权为c(T)。由于c(T′)<c(T)，可得sum(T′<sum(T)，这与T是MST 矛盾。因此，MST的瓶颈值c(T)是所有生成树中最小的，即MST必为BST。

idea：  
2.算法通过构建仅包含权重不超过 b 的边的邻接表，并利用广度优先搜索（BFS）遍历图的连通性。首先遍历所有边，筛选出权重 ≤ b 的边构建邻接表；随后从任意顶点启动 BFS，记录访问过的顶点数量，若所有顶点均被访问，则存在符合条件的生成树。  
时间复杂度：构建邻接表需遍历所有边，时间复杂度为O(E)；BFS遍历的时间复杂度为O(V+E)（每个顶点和边仅访问一次）。总时间复杂度为O(V+E)，即线性时间复杂度。

```python
# add your code here
def has_bottleneck_tree(V, E, b):
    # 构建邻接表
    adj = {v: [] for v in V}
    for u, v, w in E:
        if w <= b:
            adj[u].append(v)
            adj[v].append(u)
    
    # BFS检查连通性
    if not V:
        return True
    visited = set()
    queue = [V[0]]
    visited.add(V[0])
    while queue:
        u = queue.pop(0)
        for v in adj[u]:
            if v not in visited:
                visited.add(v)
                queue.append(v)
    return len(visited) == len(V)

V1 = [0, 1, 2]
E1 = [(0, 1, 1), (1, 2, 2), (0, 2, 3)]
b1 = 2
print(has_bottleneck_tree(V1, E1, b1))  

V2 = [0, 1, 2]
E2 = [(0, 1, 3), (1, 2, 4), (0, 2, 5)]
b2 = 2
print(has_bottleneck_tree(V2, E2, b2))  
# algorithm of the liear time complexity 
```

idea：  
3.算法通过二分法搜索最小边权阈值b，使得仅保留权重≤ b的边时图仍连通。提取所有边权重并排序，通过二分法确定最小b；每次二分使用 BFS 检查连通性；找到b后，再次遍历边集构建邻接表，并通过BFS生成树边。  
时间复杂度：提取并排序权重的时间为O(E + WlogW)（W为不同权重的数量）；二分搜索需 O(logW)次BFS检查，每次检查耗时O(E+V)，总时间为O((E+V)logW)；生成树构建中，遍历边集和BFS的时间为O(E+V)，但查找边时遍历边列表导致最坏时间O(E²)。总时间复杂度为O((E + V)log W+E²)，在W较小时为线性复杂度。

```python
def find_bottleneck_tree(V, E):
    if not V:
        return []
    
    # 检查连通性
    def is_connected(b_val):
        adj = {v: [] for v in V}
        for u, v, w in E:
            if w <= b_val:
                adj[u].append(v)
                adj[v].append(u)
        visited = set()
        if not V:
            return True
        queue = [V[0]]
        visited.add(V[0])
        while queue:
            u = queue.pop(0)
            for v in adj[u]:
                if v not in visited:
                    visited.add(v)
                    queue.append(v)
        return len(visited) == len(V)
    
    # 收集所有可能的b候选
    weights = list({w for u, v, w in E})
    weights.sort()
    left, right = 0, len(weights) - 1
    min_b = None
    while left <= right:
        mid = (left + right) // 2
        current_b = weights[mid]
        if is_connected(current_b):
            min_b = current_b
            right = mid - 1
        else:
            left = mid + 1
    if min_b is None:
        return []
    
    # 构建生成树
    adj = {v: [] for v in V}
    edges_under_b = []
    for u, v, w in E:
        if w <= min_b:
            adj[u].append(v)
            adj[v].append(u)
            edges_under_b.append((u, v, w))
    
    visited = set([V[0]])
    tree = []
    queue = [V[0]]
    while queue:
        u = queue.pop(0)
        for v in adj[u]:
            if v not in visited:
                visited.add(v)
                queue.append(v)
                # 查找对应的边
                for edge in edges_under_b:
                    if (edge[0] == u and edge[1] == v) or (edge[0] == v and edge[1] == u):
                        tree.append(edge)
                        edges_under_b.remove(edge)
                        break
    return tree

V1 = [0, 1, 2]
E1 = [(0, 1, 1), (1, 2, 2), (0, 2, 3)]
tree1 = find_bottleneck_tree(V1, E1)
print(tree1) 

V2 = [0, 1, 2, 3]
E2 = [(0, 1, 1), (1, 2, 2), (2, 3, 3), (0, 3, 4)]
tree2 = find_bottleneck_tree(V2, E2)
print(tree2) 

# algorithm of the liear time complexity 
```

## 问题 3

**道路网（Road Network）**

假设有一个以图 $ G(V, E, l) $ 表示的道路网络，连接了一组城市 $ V $。我们假设该网络是有向的，并且每条道路 $(u, v) \in E$ 都有一个非负的长度 $ l(u, v) $。一条新的道路即将被建造，因此有一个列表 $ E' $ 包含它可以连接的城市对。每对 $(u, v) \in E'$ 都有一个对应的长度 $ l'(u, v) $。我们希望选择一对城市，使得两个城市 $ s, t \in V $ 之间的距离减少最大。请为此问题编写一个高效的算法，并详细解释算法的正确性和复杂度。


idea：  
对于给定的道路网络图G(V,E,l)，目标是选择一个候选边(u,v)∈E，使得某两个城市s,t之间的最短距离减少量最大化。首先使用Floyd算法计算原图中所有节点对的最短距离矩阵，时间复杂度为O(n3)。然后遍历每个候选边(u,v)，计算添加该边后可能带来的最大距离减少量。对于任意节点对s,t，新路径可能为s→u→v→t，其长度为d_original[s][u]+l(u,v)+d_original[v][t]。若该值小于原最短距离d_original[s][t]，则减少量为原距离与新距离的差值。需遍历所有s,t找到全局最大减少量。若原距离为无穷大而新路径有效，则减少量视为无穷大。  
时间复杂度:Floyd算法的时间复杂度为O(n3)。对每个候选边(u,v)，需遍历所有s,t，时间复杂度为O(n2)。若有m条候选边，总时间为O(mn2)。  
总时间复杂度为 O(n3+mn2)。

```python
def floyd_warshall(graph):
    INF = 1 << 60  
    n = len(graph)
    dist = [[INF] * n for _ in range(n)]
    
    # 初始化距离矩阵
    for i in range(n):
        dist[i][i] = 0
        for j, w in graph[i]:
            dist[i][j] = w
    
    # Floyd-Warshall算法
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i][k] != INF and dist[k][j] != INF:
                    dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])
    return dist

def find_optimal_edge(V, E, E_prime, l_prime):
    INF = 1 << 60
    n = len(V)
    graph = [[] for _ in range(n)]
    
    # 构建原始图的邻接表
    for (u, v, l) in E:
        graph[u].append((v, l))
    
    # 计算原始图的最短路径
    d_original = floyd_warshall(graph)
    max_reduction = -INF
    best_edge = None
    
    # 遍历候选边评估最优解
    for (u, v, l_new) in E_prime:
        current_max = -INF
        
        # 判断候选边是否连通新路径
        has_infinite_reduction = False
        s_connected = any(d_original[s][u] < INF for s in range(n))
        t_connected = any(d_original[v][t] < INF for t in range(n))
        
        if s_connected and t_connected:
            for s in range(n):
                for t in range(n):
                    if d_original[s][u] < INF and d_original[v][t] < INF and d_original[s][t] == INF:
                        has_infinite_reduction = True
                        break
                if has_infinite_reduction:
                    break
        
        # 计算最大优化值
        if has_infinite_reduction:
            current_max = INF
        else:
            for s in range(n):
                for t in range(n):
                    if d_original[s][t] == INF:
                        continue
                    new_dist = d_original[s][u] + l_new + d_original[v][t]
                    if new_dist < d_original[s][t]:
                        current_max = max(current_max, d_original[s][t] - new_dist)
        
        # 更新最优边
        if current_max > max_reduction:
            max_reduction = current_max
            best_edge = (u, v, l_new)
    
    return best_edge

V = [0, 1, 2, 3] 
E = [(0, 1, 2), (1, 2, 3), (2, 3, 1)]
E_prime = [(0, 3, 4), (1, 3, 2)]  

best_edge = find_optimal_edge(V, E, E_prime, None)
print(f"最优候选边: {best_edge}")  
```

## 问题 4

**逃离问题**

一个 $ n \times n $ 的网格是一个无向图，由 $ n $ 行和 $ n $ 列的顶点组成，如下图所示。我们用 $(i,j)$ 表示第 $ i $ 行和第 $ j $ 列的顶点。除了边界顶点，网格中的所有顶点都有四个邻居，即满足 $ i = 1, i = n, j = 1 $ 或 $ j = n $ 的点 $(i,j)$。

给定网格中的 $ m \leq n^2 $ 个起点 $(x_1, y_1), (x_2, y_2), \cdots , (x_m, y_m)$，逃离问题是确定是否存在 $ m $ 条顶点不相交的路径（即路径之间不相交），从这些起点到边界上的任意 $ m $ 个不同点。例如，图1中的网格存在逃离。

(1) 该问题可以看作是一个最大流问题。考虑一个流网络，其中顶点和边都有容量。也就是说，进入任何给定顶点的总正流量受到容量限制。证明在具有边和顶点容量的网络中确定最大流可以简化为在具有可比大小的普通流网络上的最大流问题。更准确地说，你需要将一个具有顶点和边容量的网络 $ G = (V,E) $ 转换为另一个仅具有边容量的网络 $ G' = (V', E') $，使得两个网络上的最大流相同，并且你构建的新网络具有 $ V' = O(V) $ 个顶点和 $ E' = O(E) $ 条边。你可以假设网络是连通的。

(2) 描述一个解决逃离问题的高效算法，并分析其运行时间。

idea：  
1.题目：带有顶点和边容量的网络 G=(V,E)可以转换为仅含边容量的网络G′=(V′,E′)，使得两者最大流相同，且|V′|=O(|V|)，|E′|=O(|E|)。  
证明：  
给定原网络G=(V,E)，其中每个顶点v∈V 有容量c(v)≥0，每条边(u,v)∈E有容量c(u,v)≥0  
构造新网络G′=(V′,E′) ，对每个顶点v∈V，创建两个新顶点v_in和v_out，添加有向边(v_in, v_out)，容量为c(v)，对原边(u,v)∈E，添加有向边 (u_out, v_in)，容量为c(u,v)。其中，|V′|=2|V|（每个原顶点拆分为两个），|E′|=|V|+|E|（顶点拆分边 + 原边转换），满足|V′|=O(|V|) 且 |E′|=O(|E|)  
1.1.证明G的可行流⟹G′的可行流  
证明：  
设f是G的可行流，构造G′的流f′，对原边(u,v)：f′(u_out→v_in)=f(u,v)，对顶点边(v_in→v_out)：f′(v_in→v_out)=Σ_{u→v}f(u,v)，边容量f′(u_out→v_in)=f(u,v)≤c(u,v)，顶点容量：f′(v_in→v_out)=Σf(u,v)≤c(v)，由顶点拆分结构保证流量守恒  
2.2.证明：G′的可行流⟹G的可行流  
证明：  
设 f′是 G′的可行流，构造G的流f，对原边(u,v)：f(u,v)=f′(u_out→v_in)，边容量：f(u,v)=f′(u_out→v_in)≤c(u,v)，顶点容量：Σf(u,v)=f′(v_in→v_out)≤c(v)，由 G′的守恒性保证流量守恒  
所以G 与 G′ 的可行流存在双射关系，故最大流值相同。

idea：  
2.算法通过顶点拆分技术将网格中的顶点容量限制转化为边容量限制，构建了一个等效的流网络。每个网格顶点(i,j)被拆分为输入节点和输出节点，两者间的边容量为1（确保顶点仅被一条路径使用）。相邻顶点间通过输出节点到输入节点的双向边连接（容量为1，保证路径不重叠），源点连接所有起点，边界顶点连接汇点。通过计算源点到汇点的最大流，若其值等于起点数量m，则存在m条顶点不相交的逃离路径。从而将逃离问题转化为标准最大流问题，并利用Dinic算法求解。  
时间复杂度：假设网格规模为n×n，流网络的顶点数为V=2n²+2，边数E=O(n²)（每个顶点拆分产生1条边，四向邻接产生4条边/顶点）。Dinic算法在单位容量网络（所有边容量为1）中的时间复杂度为O(E√E)。代入E=O(n²)，总时间复杂度为O(n²√n²)=O(n³)。

```python
class Edge:
    def __init__(self, to, rev, cap):
        self.to = to
        self.rev = rev
        self.cap = cap

class Dinic:
    def __init__(self, n):
        self.size = n
        self.graph = [[] for _ in range(n)]
    
    def add_edge(self, fr, to, cap):
        forward = Edge(to, len(self.graph[to]), cap)
        backward = Edge(fr, len(self.graph[fr]), 0)
        self.graph[fr].append(forward)
        self.graph[to].append(backward)
    
    def bfs_level(self, s, t, level):
        q = []
        level[:] = [-1] * self.size
        level[s] = 0
        q.append(s)
        while q:
            v = q.pop(0)
            for edge in self.graph[v]:
                if edge.cap > 0 and level[edge.to] == -1:
                    level[edge.to] = level[v] + 1
                    q.append(edge.to)
                    if edge.to == t:
                        return
    
    def dfs_flow(self, v, t, flow, level, ptr):
        if v == t:
            return flow
        while ptr[v] < len(self.graph[v]):
            edge = self.graph[v][ptr[v]]
            if edge.cap > 0 and level[v] < level[edge.to]:
                min_flow = min(flow, edge.cap)
                result = self.dfs_flow(edge.to, t, min_flow, level, ptr)
                if result > 0:
                    edge.cap -= result
                    self.graph[edge.to][edge.rev].cap += result
                    return result
            ptr[v] += 1
        return 0
    
    def max_flow(self, s, t):
        flow = 0
        level = [-1] * self.size
        while True:
            self.bfs_level(s, t, level)
            if level[t] == -1:
                return flow
            ptr = [0] * self.size
            while True:
                f = self.dfs_flow(s, t, float('inf'), level, ptr)
                if f == 0:
                    break
                flow += f
            level = [-1] * self.size

def solve_escape(n, starts):
    total_nodes = 2 + 2 * n * n
    dinic = Dinic(total_nodes)
    S, T = 0, 1
    
    def get_in_id(i, j):
        return 2 + 2 * ((i-1) * n + (j-1))
    
    # 顶点拆分
    for i in range(1, n+1):
        for j in range(1, n+1):
            in_node = get_in_id(i, j)
            dinic.add_edge(in_node, in_node + 1, 1)
    
    # 邻接边
    dirs = [(-1,0), (1,0), (0,-1), (0,1)]
    for i in range(1, n+1):
        for j in range(1, n+1):
            u_out = get_in_id(i, j) + 1
            for di, dj in dirs:
                ni, nj = i + di, j + dj
                if 1 <= ni <= n and 1 <= nj <= n:
                    v_in = get_in_id(ni, nj)
                    dinic.add_edge(u_out, v_in, 1)
    
    # 源点连接起点
    for x, y in starts:
        dinic.add_edge(S, get_in_id(x, y), 1)
    
    # 边界连接汇点
    for i in range(1, n+1):
        for j in range(1, n+1):
            if i == 1 or i == n or j == 1 or j == n:
                dinic.add_edge(get_in_id(i, j)+1, T, 1)
    
    return dinic.max_flow(S, T) == len(starts)

n = 6
starts = [(3,1), (2,2), (3,2), (4,2), (2,4), (3,4), (4,4), (2,6), (3,6), (4,6)]
print(solve_escape(n, starts))  
```
